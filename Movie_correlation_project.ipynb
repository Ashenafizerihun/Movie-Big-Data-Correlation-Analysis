```python
#import libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')

from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12, 8)

#Read date
file_id = "18LgIi9aCQDD5ZlGLlEIwJoeKYfDw_XFG"
direct_link = f"https://drive.google.com/uc?id={file_id}"
df = pd.read_csv(direct_link)
```

```python
#Drope some columns
df.drop(['index','homepage','id','keywords','original_language','spoken_languages','tagline','cast','crew','director','original_title'],axis=1,inplace=True)
```

```python
pd.set_option('display.max_rows', 10)  # Display only 10 rows
```

```python
df.info()
```

```python
#Drop null runtime value rows
df.dropna(subset=['runtime'], inplace=True)
df['runtime'] = df['runtime'].astype('int64')
```

```python
df
```

```python
#Sort the dataframe by revenue column in descending order
df.sort_values(by=['revenue'],inplace=True,ascending=False)
```

```python
df
```

```python
#Drop duplicates throughout the dataframe
df.drop_duplicates()
```

```python
# Scatter plot of budget Vs revenue
plt.scatter(df['budget'],df['revenue'])
plt.title('Budget Vs revunue')
plt.xlabel('budget')
plt.ylabel('revenue')
```

```python
# Budget Vs Revenue regression plot
sns.regplot(x='budget',y='revenue',data=df,scatter_kws={'color':'red'},line_kws={'color':'black'})
```

```python
# corelation between numeric columns
df.corr(numeric_only=True)
```

```python
# heatmap between movie variables
correlation_matrix = df.corr(numeric_only=True)
sns.heatmap(correlation_matrix,annot=True)
plt.title('Heatmap')
```

```python
df.info()
```

```python
# Convert object columns to 'category' and numerize it
df_numerized = df.copy()
for df_column in df_numerized.columns:
    if df_numerized[df_column].dtype == 'object':  # Check if the column is of type 'object'
        df_numerized[df_column] = df_numerized[df_column].astype('category')  # Convert to category
        df_numerized[df_column] = df_numerized[df_column].cat.codes
df = df_numerized
df
```

```python
# heatmap between movie variables
correlation_matrix = df_numerized.corr(numeric_only=True)
sns.heatmap(correlation_matrix,annot=True)
plt.title('Heatmap')
```

```python
# coorelation between movie variables unstacked
correlation_matrix = df.corr()
correlation_matrix = correlation_matrix.unstack()

with pd.option_context('display.max_rows', None):
    print(correlation_matrix)
```
